# Task 2 (3 points)

Split data into train and test. Set `test_size=0.3, random_state=0`

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)
```

Preprocess the data:


*   Depict number of missing values in every column, replace missing values with median, computed on train data (numerical features) or with 'Not Given' (categorical features)
*   Encode the categorical with one-hot-encoder



```python
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object', 'bool']).columns

values = {
    **{col: X_train[col].median() for col in numerical_features},
    **{col: 'Not Given' for col in categorical_features},
}
X_test.fillna(values, inplace=True)
X_train.fillna(values, inplace=True)

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False) 
X_train_encoded = encoder.fit_transform(X_train[categorical_features])
X_test_encoded = encoder.transform(X_test[categorical_features])

X_train_final = pd.concat([
    pd.DataFrame(X_train[numerical_features]),
    pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out())
], axis=1)

X_test_final = pd.concat([
    pd.DataFrame(X_test[numerical_features]),
    pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out())
], axis=1)
```

Train linear regression without regularization, compute MSE on train and test data.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

X_train_final = X_train_final.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
X_test_final = X_test_final.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

train_data = pd.concat([X_train_final, y_train], axis=1)
test_data = pd.concat([X_test_final, y_test], axis=1)

train_cleaned = train_data.dropna()
test_cleaned = test_data.dropna()

X_train_clean = train_cleaned.drop(columns=[y_train.name])
y_train_clean = train_cleaned[y_train.name]
X_test_clean = test_cleaned.drop(columns=[y_test.name])
y_test_clean = test_cleaned[y_test.name]

print(f"Осталось train образцов: {len(X_train_clean)}")
print(f"Осталось test образцов: {len(X_test_clean)}")

if len(X_train_clean) > 10 and len(X_test_clean) > 10:  
    model = LinearRegression()
    model.fit(X_train_clean, y_train_clean)
    
    y_train_pred = model.predict(X_train_clean)
    y_test_pred = model.predict(X_test_clean)
    
    mse_train = mean_squared_error(y_train_clean, y_train_pred)
    mse_test = mean_squared_error(y_test_clean, y_test_pred)
    
    print(f"\nMSE Train: {mse_train:.2f}")
    print(f"MSE Test: {mse_test:.2f}")
```

# Task 3 (3 points)

For different `alpha` hyperparameter fit linear regression with $L_1$ regularization (`Lasso` model, if there are warnings - set `tol=0.1`).

Choose the grid in such way that the smallest alpha will result in 0-5 zero coefficent and for biggest - all coefficient will be 0.

For every model save number of zero coefficients and MSE on train and test data

```python
from sklearn.linear_model import Lasso
import numpy as np

# Подбираем диапазон alpha так, чтобы:
# - Минимальный alpha давал 0-5 нулевых коэффициентов
# - Максимальный alpha обнулял все коэффициенты
alphas = np.logspace(-3, 2, 50)  # 50 значений от 0.001 до 100

results = []
for alpha in alphas:
    model = Lasso(alpha=alpha, tol=0.1, random_state=0)
    model.fit(X_train_clean, y_train_clean)
    
    # Количество нулевых коэффициентов
    n_zero = sum(model.coef_ == 0)
    
    # MSE на train и test
    mse_train = mean_squared_error(y_train_clean, model.predict(X_train_clean))
    mse_test = mean_squared_error(y_test_clean, model.predict(X_test_clean))
    
    results.append({
        'alpha': alpha,
        'n_zero': n_zero,
        'mse_train': mse_train,
        'mse_test': mse_test
    })

# Конвертируем в DataFrame для удобства
results_df = pd.DataFrame(results)
```

Draw 2 plots near each other: plot of number of aero coefficients and plot of MSE (both train and test).  Draw conclusions: compare the MSE of Lasso and initial linear regression, which features are considered important in the best model, is it logical.

```python
import matplotlib.pyplot as plt

# Создаем фигуру с двумя графиками
plt.figure(figsize=(15, 5))

# График 1: Количество нулевых коэффициентов
plt.subplot(1, 2, 1)
plt.plot(results_df['alpha'], results_df['n_zero'], 'b-o')
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('Number of zero coefficients')
plt.title('Feature Sparsity vs Alpha')

# График 2: MSE на train и test
plt.subplot(1, 2, 2)
plt.plot(results_df['alpha'], results_df['mse_train'], 'r-', label='Train MSE')
plt.plot(results_df['alpha'], results_df['mse_test'], 'g--', label='Test MSE')
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('MSE')
plt.legend()
plt.title('Model Performance vs Alpha')

plt.tight_layout()
plt.show()

# Анализ лучшей модели (с минимальным test MSE)
best_idx = results_df['mse_test'].idxmin()
best_alpha = results_df.loc[best_idx, 'alpha']
best_model = Lasso(alpha=best_alpha, tol=0.1, random_state=0)
best_model.fit(X_train_clean, y_train_clean)
```

Important features in the best Lasso model (alpha=0.176):
ID, Previous Scores (out of 10), Hours Studied, Hours Free time, Sleep Hours, Weight, BMI, Eye color_brown

MSE Comparison:
Lasso (Test): 37.94
Linear (Test): 41.39

Conclusion: Lasso regression with alpha=0.176 gives comparable accuracy
to conventional linear regression, but uses only 8 features.
The selection of features is logical: the most prognostically significant
features are preserved, such as 'Previous Scores' and 'Hours Studied'.

# Task 4 (2 points)

Now, use $L_2$ regularization. Tune `Ridge` model hyperparameters using cross-validation. Report the best MSE.  

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import numpy as np

params = {'alpha': np.logspace(-12, 10, 50)}

ridge = Ridge()
grid_search = GridSearchCV(ridge, params, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_clean, y_train_clean)

best_ridge = grid_search.best_estimator_
best_alpha = grid_search.best_params_['alpha']

y_train_pred = best_ridge.predict(X_train_clean)
y_test_pred = best_ridge.predict(X_test_clean)
mse_train = mean_squared_error(y_train_clean, y_train_pred)
mse_test = mean_squared_error(y_test_clean, y_test_pred)

print(f"Best alpha: {best_alpha:.2e}")
print(f"MSE Train: {mse_train:.2f}")
print(f"MSE Test: {mse_test:.2f}")
```

Investigate coefficients: how many zeros? Which features are important?

```python
coef_df = pd.DataFrame({
    'Feature': X_train_clean.columns,
    'Coefficient': best_ridge.coef_,
    'Abs_Coef': np.abs(best_ridge.coef_)
}).sort_values('Abs_Coef', ascending=False)

# Визуализация топ-10 коэффициентов
plt.figure(figsize=(10, 6))
plt.barh(coef_df['Feature'][:10], coef_df['Coefficient'][:10])
plt.title('Top 10 important signs (Ridge)')
plt.xlabel('Coefficient value')
plt.tight_layout()
plt.show()

n_zero = sum(best_ridge.coef_ == 0)
print(f"\nNumber of zero coefficients: {n_zero}")
```
